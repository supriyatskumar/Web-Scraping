```python
from selenium import webdriver
import numpy as np
import pandas as pd
import datetime
import requests
import re
```

```python
options = webdriver.ChromeOptions()

options.add_argument("headless")

browser = webdriver.Chrome(options=options)

url = "https://www.google.com/maps/place/Hall+Post+1125/@42.4074567,-82.9576057,17z/data=!4m7!3m6!1s0x8824d42c03461229:0x8f1e765b328511f5!8m2!3d42.4074567!4d-82.955417!9m1!1b1"

browser.get(url)
```

```python
review_titles = browser.find_elements_by_class_name("section-review-title")
users=[a.text for a in review_titles]

review_text = browser.find_elements_by_class_name("section-review-review-content")
reviews=[a.text for a in review_text]

review_stars = browser.find_elements_by_class_name("section-review-stars")
ratings=[len(a.find_elements_by_class_name("section-review-star-active")) for a in review_stars]

review_date=browser.find_elements_by_class_name("section-review-publish-date")
relative_date=[a.text for a in review_date]
```

```python
count=browser.find_elements_by_class_name("gm2-caption")
cnt=[a.text for a in count]
review_total=cnt[1]
Current_date=datetime.datetime.today().strftime('%m-%d-%Y')
```

```python
print('Data collection date: ',Current_date,'\nTotal Reviews: ',review_total)
```

```python
len(users),len(reviews),len(ratings),len(relative_date)
```

```python
index=8
for x in range((int(review_total[:2])-8)%10):
    req='https://www.google.com/maps/preview/review/listentitiesreviews?authuser=0&hl=en&gl=us&pb=!1m2!1y9810199173818356265!2y10312810330786697717!2m2!1i'+str(index)+'!2i10!3e1!4m5!3b1!4b1!5b1!6b1!7b1!5m2!1sjsRGXonvGMCO9PwPzbe16Ao!7e81'
    r=requests.get(req)
    review_author=re.findall('u003den-US",(.+?),"https://lh',r.text)
    t=re.findall(r',"(.+)https://www.google.com/local/review/rap/redirect',r.text)
    review_rate=[t[i].split(',')[3] for i in range(len(t))]
    review_content=[t[i].split(',')[2] for i in range(len(t))]
    review_relative_date=[t[i].split(',')[0] for i in range(len(t))]
    for n in range(len(review_relative_date)):
        users.append(review_author[n][1:-1])
        reviews.append(review_content[n])
        ratings.append(review_rate[n])
        relative_date.append(review_relative_date[n][:-1])
    index=index+10
```
