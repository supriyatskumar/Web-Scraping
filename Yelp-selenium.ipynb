```python
review_titles = browser.find_elements_by_class_name("text-weight--bold__373c0__3HYJa")
users=[a.text for a in review_titles]
#print('1:',[a.text for a in review_titles])

review_text = browser.find_elements_by_class_name("lemon--span__373c0__3997G")
reviews=[a.text for a in review_text]
#print('2:',[a.text for a in review_text])

review_stars = browser.find_elements_by_class_name("overflow--hidden__373c0__3Usf-")
ratings=[a.text for a in review_stars]
#print('3:',[len(a.find_elements_by_class_name("section-review-star-active")) for a in stars])

review_date=browser.find_elements_by_class_name("text-align--left__373c0__2pnx_")
relative_date=[a.text for a in review_date]
```
